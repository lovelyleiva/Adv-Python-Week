```python
#we want tot get the HTML from the website in order to read it.
import requests

#We want python to understand the HTML
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://www.countrycode.org/')
html = response.text #this is the html of the website
```

```python
#make the text mean something to python
soup = BeautifulSoup(html,'html.parser') #the thing you want to interpret and how you want to interpret
#the point of soup is python knows what html does. it knows title, head, tag, etc
```

```python
#strategy: go through each of the rows in a table. For loop
#table must be the tbody
#tr must represent the row of each table
#td must represent each individual cell

#for row in table, 
#   grab each thing

#for row in table_rows:
#   table_datas = row.find_all('td')
```

```python
table_body = soup.find('tbody')
table_rows = table_body.find_all('tr')
table_datas = table_rows[0].find_all('td')
table_datas[0].string
```

```python
countries = []

for row in table_rows:
    table_datas = row.find_all('td')
    country = {}
    country['name'] = table_datas[0].string
    country['country code'] = table_datas[1].string
    countrycodes = table_datas[2].string.split('/')
    country['iso-2'] = countrycodes[0]
    country['iso-3'] = countrycodes[1]
    country['population']=table_datas[3].string
    country['Area'] = table_datas[4].string
    country['GDP'] = table_datas[5].string
    countries.append(country)

list(countries)
```
